```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Carga do dataset
#zoo = pd.read_csv('/content/drive/My Drive/_Aprendiz_Máquina/zoo.csv')
cancer = pd.read_csv('breast-cancer.csv',',')
cancer.head()
```

```python
#Separa o conjunto de dados em treino e teste
#Eliminate the first and last attributes
X = cancer.iloc[:, 2:-1].values  

y = cancer.iloc[:, 1].values  
print(X.shape)
```

# Execução do kNN
## Método de amostragem de Cross-Validation

```python
#Cross Validation
neigh = KNeighborsClassifier(n_neighbors=3,weights='distance')
predictions_train = cross_val_predict(neigh,X,y,cv=10)
```

```python

accuracy_score(y,predictions_train)*100
```

```python
#Gera a matriz de confusão dos dados de treino
confusion_matrix(y,predictions_train)
```

```python
#Gera a matriz de confusão do treino na visualização de HeatMMap
import seaborn as sns
cf = confusion_matrix(y,predictions_train)
lbl1=['M', 'B']
lbl2 = ['M', 'B']
sns.heatmap(cf,annot=True,cmap="Greens", fmt="d",xticklabels=lbl1,yticklabels=lbl2)
```

### Caso 1 => dados aleatoriamente gerados

```python
# Consulte a classe make_blobs
#https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html
from sklearn.datasets import make_blobs
# Xnew contem novas instâncias para as quais não se conhece a classe
Xnew, _ = make_blobs(n_samples=10, centers=4, n_features=29, random_state=1)
# realiza a predição
neigh = KNeighborsClassifier(n_neighbors=3,weights='distance')
neigh.fit(X, y)
ynew = neigh.predict(Xnew)
# exibe as entradas e a saída predita
for i in range(len(Xnew)):
	print("X=%s, Predicted=%s" % (Xnew[i], ynew[i]))
```
