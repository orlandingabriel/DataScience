```python
# importa o dataset iris 
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# cria dataset iris 
iris = load_iris()
```

```python

# Extrai valores de dados (atributos/features) e cria a lista de instâncias 
featuresAll=[]
features = iris.data[: , [0,1,2,3]]
features.shape
```

```python
# Extrai as classes (targets)
targets = iris.target
targets.reshape(targets.shape[0],-1)
targets.shape
print(iris.target)
```

```python
# Escreve os atributos/features (menos o atributo classificador)
featuresAll=[]
for observation in features:
    featuresAll.append([observation[0] + observation[1] + observation[2] + observation[3]])
print (featuresAll[:4])
```

```python
# Plota as instâncias
plt.scatter(featuresAll, targets, color='blue', alpha =1.0)
plt.rcParams['figure.figsize'] = [10,4]
plt.title('Dataset Iris - Scatter Plot')
plt.xlabel('Atributos')
plt.ylabel('Classes')
```

```python
#Scatter Plot do Iris Dataset (Relação entre Comprimento da Sepala e Largura da Sepala) 
#Busque a Relação entre Comprimento da Sepala e Largura da Sepala
featuresAll = []
targets = []
for feature in features:
    featuresAll.append(feature[0]) #comprimento da Sepala 
    targets.append(feature[1]) #Largura da sepala 

groups = ('Iris-setosa','Iris-versicolor','Iris-virginica')
colors = ('blue', 'green','red')
data = ((featuresAll[:50], targets[:50]), (featuresAll[50:100], targets[50:100]), 
        (featuresAll[100:150], targets[100:150]))

for item, color, group in zip(data,colors,groups): 
    #item = (featuresAll[:50], targets[:50]), (featuresAll[50:100], targets[50:100]), (featuresAll[100:150], targets[100:150])
    x, y = item
    plt.scatter(x, y,color=color,alpha=1)
    plt.title('Iris Dataset Scatter Plot')
   
plt.xlabel('Comprimento da Sépala')
plt.ylabel('Largura da Sépala')
plt.show()

#scatter =  plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, )
#plt.legend(handles=scatter.legend_elements()[0], labels=list(iris.target_names))
```

```python
#Scatter Plot do Iris Dataset (Relação entre Comprimento da Pétala e Largura da Pétala) 
featuresAll = []
targets = []
for feature in features:
    featuresAll.append(feature[2]) #Comprimento da Pétala
    targets.append(feature[3]) #Largura da Pétala

groups = ('Iris-setosa','Iris-versicolor','Iris-virginica')
colors = ('blue', 'green','red')
data = ((featuresAll[:50], targets[:50]), (featuresAll[50:100], targets[50:100]), 
        (featuresAll[100:150], targets[100:150]))

for item, color, group in zip(data,colors,groups): 
    #item = (featuresAll[:50], targets[:50]), (featuresAll[50:100], targets[50:100]), (featuresAll[100:150], targets[100:150])
    x, y = item
    plt.scatter(x, y,color=color,alpha=1)
    plt.title('Iris Dataset scatter Plot')
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Largura da Pétala')
plt.legend()
plt.show()
```
