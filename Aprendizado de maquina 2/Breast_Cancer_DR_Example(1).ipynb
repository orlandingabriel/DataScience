# Breast Cancer dataset
The Breast Cancer data set is a real-valued multivariate data that consists of two classes, where each class signifies whether a patient has breast cancer or not. The two categories are: malignant and benign.
The malignant class has 212 samples, whereas the benign class has 357 samples.
It has 30 features shared across all classes: radius, texture, perimeter, area, smoothness, fractal dimension, etc.

```python
#import dataset from sklearn's module datasets
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
breast = load_breast_cancer()
breast_data = breast.data
breast_data.shape
```

```python
breast_labels = breast.target
breast_labels.shape
```

```python
#reshape labels
labels = np.reshape(breast_labels,(569,1))
#concatenate the data and labels
final_breast_data = np.concatenate([breast_data,labels],axis=1)
final_breast_data.shape
```

```python
#convert to a dataframe to represent the data in a table
breast_dataset = pd.DataFrame(final_breast_data)
breast_dataset.head()
```

```python
features = breast.feature_names
#missing target field
features
```

```python
features_labels = np.append(features,'target')
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
#replace target field
breast_dataset['target'].replace(0, 'Benign',inplace=True)
breast_dataset['target'].replace(1, 'Malignant',inplace=True)
```

```python
breast_dataset.tail()
```

```python
df = breast_dataset.drop('target',1)
df.corr()
```

```python
df.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1, max_depth=10)
model.fit(df,breast_dataset.target)
```

```python
features = df.columns
importances = model.feature_importances_
#indexes from the top 10 importances fetures
indices = np.argsort(importances)[-10:]  # top 10 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn import datasets
lreg = LinearRegression()
rfe = RFE(lreg, 10)
rfe.fit_transform(df,breast_dataset.target)
```

```python
features = df.columns
importances_reg = rfe.ranking_
indices_reg = np.argsort(importances_reg)  # top 10 features
plt.figure(figsize=(7,7))
plt.title('Feature Importances')
plt.barh(range(len(indices_reg)), importances_reg[indices_reg], color='b', align='center')
plt.yticks(range(len(indices_reg)), [features[i] for i in indices_reg])
plt.xlabel('Relative Importance')
plt.show()
```

```python
from sklearn.feature_selection import f_regression
#output: 1 array of F-values and another array of p-values
ffs = f_regression(df,breast_dataset.target)
```

```python
#Take only f-values
ffs_f_values = ffs[0]
indices_forw = np.argsort(ffs[0])
plt.figure(figsize=(7,7))
plt.title('Feature Importances')
plt.barh(range(len(indices_forw)), ffs_f_values[indices_forw], color='b', align='center')
plt.yticks(range(len(indices_forw)), [features[i] for i in indices_forw])
plt.xlabel('Relative Importance')
plt.show()
```

```python
variable = [ ]
for i in range(0,len(df.columns)-1):
    #take F-values greater than 10
    if ffs[0][i]>=600:
        variable.append(df.columns[i])
print(variable)
```

# PCA Dimension Reduction Visualization

```python
##Data visualization using PCA
#Standardizing the data since PCA's output is influenced based on the scale of the features of the data
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # normalizing the features
```

```python
x.shape
```

```python
#check the mean of zero and a standard deviation of one on the normalized data
np.mean(x),np.std(x)
```

```python
#give column names
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
#convert to a Dataframe
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
#Create 2 PC components
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principal_breast_Df.tail()
```

```python
principal_breast_Df.describe()
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib.pyplot as plt
per_var = np.round(pca_breast.explained_variance_ratio_ * 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var) + 1)]
plt.bar(x=range(1, len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('percentange of explained variance')
plt.xlabel('principal component')
plt.title('Pincipal components ratio')
plt.show()
```

From the above output, you can observe that the principal component 1 holds 44.2% of the information while the principal component 2 holds only 19% of the information. Also, the other point to note is that while projecting thirty-dimensional data to a two-dimensional data, 36.8% information was lost.

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['target'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
