```python
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import Perceptron
import sklearn.metrics as metric
import numpy as np
```

```python
X_treino=[[1, 1], 
            [1, 0],
            [0, 1],
            [0, 0]
           ]
y_treino=[1, 
            1,
            1,
            0
           ]
X_teste=X_treino
y_teste=y_treino
```

# Perceptron

```python
ptn = Perceptron(max_iter=50,tol=0.001)                    
ptn.fit(X_treino, y_treino)                    # treino
y_pred=ptn.predict(X_teste)                      # predição
print(y_pred)                                      # saída
accuracy=metric.accuracy_score(y_teste, y_pred, normalize=True)
print('acurácia=',accuracy)                         
```

```python
print(ptn.intercept_, ptn.coef_) 
```

# Rede Neural MLP


```python
mlp = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(1,1), activation='logistic') # parâmetros da rede neural
mlp.fit(X_treino, y_treino)                        # treino
y_pred=mlp.predict(X_teste)                        # predição
print(y_pred)                                      # apresenta a saída
accuracy=metric.accuracy_score(np.array(y_teste).flatten(), np.array(y_pred).flatten(), normalize=True)
print('acuracy=',accuracy)                      
```

```python
print([coef.shape for coef in mlp.coefs_])  # arrays dos pesos sinápticos
mlp.coefs_                                  # pesos sinápticos 
```
